**Let's create a simple short Binary Tree and print it , starting with creating the nodes then linking the nodes to establish the relationships between nodes and then printing it in a descriptive manner**

```python
class BinaryTreeNode:
    
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
```

```python
# Creating the Nodes for binary trees
btn1 = BinaryTreeNode(1)
btn2 = BinaryTreeNode(2)
btn3 = BinaryTreeNode(3)
btn4 = BinaryTreeNode(4)
btn5 = BinaryTreeNode(5)
```

```python
# Linking the nodes

# Root node
btn1.left = btn2
btn1.right = btn3
#subtrees
btn2.left = btn4
btn2.right = btn5
```

```python
# Printing the Binary Tree using Recursion

def PrintBT(root):
    if root is None:
        return
    print(root.data, end=': ')
    if root.left is not None:
        print('L',  root.left.data, end= ',')
    if root.right is not None:
        print('R', root.right.data, end=' ')
    print()
    PrintBT(root.left)
    PrintBT(root.right)
```

```python
PrintBT(btn1)
```

**TAKING INPUT FROM THE USER**

```python
def takeinputBT():
    root_node = int(input())
    if root_node == -1:
        return
    root = BinaryTreeNode(root_node)
    left_tree = takeinputBT()
    right_tree = takeinputBT()
    root.left = left_tree
    root.right = right_tree
    return root
```

**Writing code for counting No. of nodes in BT**

```python
def NumNodes(root):
    if root is None:
        return 0
    left_count = NumNodes(root.left)
    right_count = NumNodes(root.right)
    
    return 1 + left_count + right_count
```

```python
root = takeinputBT()
PrintBT(root)
NumNodes(root)
```

**Let's see if we could write a program for Finding sum for all nodes**

```python
def sumOfAllNodes(root):
    if root is None:
        return 0
    leftSum = sumOfAllNodes(root.left)
    rightSum = sumOfAllNodes(root.right)
    return root.data + leftSum + rightSum

sumOfAllNodes(root)
```

**We will write code for types of Tree Traversing: Pre-Order, Post-Order and In-Order. Concepts are written in Notes**

```python
# Pre Order
def preorder(root):
    if root is None:
        return
    print(root.data, end=' ')
    preorder(root.left)
    preorder(root.right)
    return

preorder(root)
```

```python
# Post Order
def postorder(root):
    if root is None:
        return
    postorder(root.left)
    postorder(root.right)
    print(root.data, end=' ')
    return

postorder(root)
```

**Node with largest data**

```python
def largestdata(root):
    if root is None:
        return -1
    left= largestdata(root.left)
    right = largestdata(root.right)
    return max(root.data, left, right)

largestdata(root)
```

**Nodes Greater than x**

```python
def countNodesGreaterThanX(root, x):
    if root is None:
        return 0
    leftcount = countNodesGreaterThanX(root.left, x)
    rightcount = countNodesGreaterThanX(root.right, x)
    if root.data > x:
        return 1 + leftcount + rightcount
    return leftcount + rightcount

countNodesGreaterThanX(root, 2)
```

**Height of Tree**

```python
def height(root):
    if root is None:
        return 0
    leftcount = height(root.left)
    rightcount = height(root.right)
    return max(leftcount, rightcount) + 1

height(root)
```

**Number Of Leaf Nodes**

```python
def leafNodes(root):
    if root is None:
        return 0
    if root.left is None and  root.right is None:
        return 1
    return leafNodes(root.left) + leafNodes(root.right)

leafNodes(root)
```

**Print Nodes At Depth K**

```python
def PrintAtK(root, k):
    if root is None:
        return
    if k == 0:
        print(root.data)
        return
    PrintAtK(root.left, k-1)
    PrintAtK(root.right, k-1)
    
PrintAtK(root, 2)
```

**Is node present ?**

```python
def isNodePresent(root, x):
    if root is None:
        return
    if root.data == x:
        return True
    leftAns=  isNodePresent(root.left, x)
    rightAns= isNodePresent(root.right, x)
    return leftAns or rightAns

isNodePresent(root, 4)
```

**Nodes without sibling**

```python
def nodesWithoutSibling(root):
    if root is None:
        return
    if root.left is None and root.right is None:
        return
    if root.left is None and root.right is not None:
        print(root.right.data)
        nodesWithoutSibling(root.right)
    if root.right is None and root.left is not None:
        print(root.left.data)
        nodesWithoutSibling(root.left)
    if root.left and root.right:
        nodesWithoutSibling(root.left)
        nodesWithoutSibling(root.right)  
        
nodesWithoutSibling(root)
```
