```python
class BinaryTreeNode:
    
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
```

```python
# TAKING INPUT FROM THE USER

def takeinputBT():
    root_node = int(input())
    if root_node == -1:
        return
    root = BinaryTreeNode(root_node)
    left_tree = takeinputBT()
    right_tree = takeinputBT()
    root.left = left_tree
    root.right = right_tree
    return root
```

```python
# Printing the Binary Tree using Recursion

def PrintBT(root):
    if root is None:
        return
    print(root.data, end=': ')
    if root.left is not None:
        print('L',  root.left.data, end= ',')
    if root.right is not None:
        print('R', root.right.data, end=' ')
    print()
    PrintBT(root.left)
    PrintBT(root.right)
    
root = takeinputBT()
PrintBT(root)
```

   ** Remove the leaf Nodes**

```python
def removeleaf(root):
    if root is None:
        return
    if root.left is None and root.right is None:
        return None
    root.left = removeleaf(root.left)
    root.right = removeleaf(root.right)
    return root

removeleaf(root)
PrintBT(root)
```

**Mirror a binary Tree: mirror Nodes, and not just data**

```python
def mirrorBinaryTree(root):
    if root is None:
        return
    left = root.left
    root.left = root.right
    root.right = left
    mirrorBinaryTree(root.left)
    mirrorBinaryTree(root.right)

root = takeinputBT()
PrintBT(root)
mirrorBinaryTree(root)
PrintBT(root)
```

**Check if the Tree is Balanced or not**

```python
def height(root):
    if root is None:
        return 0
    return 1 + height(root.left) + height(root.right)

def isBalanced(root):
    if root is None:
        return True
    left_height = height(root.left)
    right_height = height(root.right)
    if (left_height - right_height) > 1 or (right_height - left_height) > 1:
        return False
   
    if isBalanced(root.left) and isBalanced(root.right):
        return True
    return False
    
root = takeinputBT()
PrintBT(root)
isBalanced(root)
```

*The Above solution is a bad solution, its time complexity is greater than O(n) in any case: why?
because we are calling the height funtion 2 times at every call, first for finding the height itself and then calling the recursive fun isbalanced that inturns finds height again*

*We will write a better fun that in one go, while calculating and returning the height also returns whether subtree is balanced or not, along with it*

```python
def  checkBalanced(root):
    if root is None:
        return 0, True    # return base values for both funtions right here.
    
    lh, leftBalanced = checkBalanced(root.left)
    rh, rightBalanced = checkBalanced(root.right)
    
    h = 1 + max(lh, rh)
    
    # make sure your recursive fun also returns 2 things at every call since you asked for it right from base case
    if (lh - rh) > 1 or (rh - lh) > 1:
        return h, False
    
    if leftBalanced and rightBalanced:
        return h, True
    
    return h, False

def isBalanced2(root):
    h, isTreeBalanced = checkBalanced(root)
    return isTreeBalanced

root = takeinputBT()
PrintBT(root)
isBalanced2(root)
    
```

**Check the diameter of the Binary Tree**

```python
dia, ht = 0, 0
def diameter(root):
    if root is None:
        return dia, ht
    
    lelf_dia , left_ht = diameter(root.left)
    right_dia, right_ht = diameter(root.right)
    ht = 1 + max(left_ht, right_ht)
    
    # 3 possible measurement for diamater depending upon the structure of tree
    measure1 = 1 + left_ht + right_ht           # that's diamter of tree with the root node
    measure2 = left_dia
    measure3 = right_dia
    
    # Correct diameter will me indeed maximum of those three.
    return max(measure1, max(measure2, measure3)), ht
    
dia, h = diameter(root)
print(dia)
```

```python
def printLevelWise(root):
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        currNode = q.get()
        print(currNode.data,end=':')
        if currNode.left is not None:
            print('L:',end="")
            print(currNode.left.data,end=",")
            q.put(currNode.left)
        if currNode.left is None:
            print('L:-1', end=',')
        if currNode.right is not None:
            print('R:',end='')
            print(currNode.right.data,end="")
            q.put(currNode.right)
        if currNode.right is None:
            print('R:-1', end='')
        print()    
```

**Construct Tree Using Inorder and PostOrder**

```python
def buildTreePostOrder(postorder, inorder):
    if len(postorder) < 1:
        return None
    
    root = postorder[-1]
    rootNode = BinaryTreeNode(root)
    
    rootIndex_Inorder = inorder.index(root)
    left_inorder = inorder[: rootIndex_Inorder]
    right_inorder = inorder[rootIndex_Inorder+1 :]
    
    left_postorder = postorder[: rootIndex_Inorder]
    right_postorder = postorder[rootIndex_Inorder: -1]
    
    leftTree = buildTreePostOrder(left_postorder, left_inorder)
    rightTree = buildTreePostOrder(right_postorder, right_inorder)
    
    rootNode.left = leftTree
    rootNode.right = rightTree
    return rootNode
```

**Construct Tree Using Inorder and Preorder**

```python
def buildTreePreOrder(preorder, inorder):
    if len(preorder) < 1:
        return None
    root = preorder[0]
    rootNode = BinaryTreeNode(root)
    
    root_in_inorder = inorder.index(root)
    left_io = inorder[:root_in_inorder]
    right_io = inorder[root_in_inorder + 1:]
    
    len_left_io = len(left_io)
    left_preO = preorder[1: (len_left_io + 1)]
    right_preO = preorder[(len_left_io + 1): ]
    
    leftTree = buildTreePreOrder(left_preO, left_io)
    rightTree = buildTreePreOrder(right_preO, right_io)
    
    rootNode.left = leftTree
    rootNode.right = rightTree
    return rootNode
```

**Create & Insert Duplicate Node: *Using Queues***

```python
def insertDuplicateNode(root):
    if root is None:
        return
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        currNode = q.get()
        if currNode.left != None:
            q.put(currNode.left)
        if currNode.right != None:
            q.put(currNode.right)
        if currNode.left == None and currNode.right == None:
            currNodeCopy = BinaryTreeNode(currNode.data)
            currNode.left = currNodeCopy
        else:
            temp = currNode.left
            currCopy = BinaryTreeNode(currNode.data)
            currNode.left = currCopy
            currCopy.left = temp
    return root
```

**Create & Insert Duplicate Node: *Using Recursion***

```python
def insertDuplicateNode2(root):
    if root is None:
        return
    copyNode = BinaryTreeNode(root.data)
    copyNode.left = root.left
    root.left = copyNode
    insertDuplicateNode2(copyNode.left)
    insertDuplicateNode2(root.right)
```

**Path Sum Root to Leaf**

```python
def rootToLeafPathsSumToK(root, k, lst):
    if root is None:
        return
    if root.left is None and root.right is None:
        if root.data == k:
            print(*lst, k)
    lst.append(root.data)
    rootToLeafPathsSumToK(root.left, k-root.data, lst)
    rootToLeafPathsSumToK(root.right, k-root.data, lst)
    lst.pop()
    
```

**Given a binary tree, find and return the min and max data value of given binary tree.**

```python
INT_MIN = -2147483648
INT_MAX = 2147483647

def minMax(root):
    if root is None:
        return INT_MAX, INT_MIN
    if root.left is None and root.right is None:
        return root.data, root.data
    left_min, left_max = minMax(root.left)
    right_min, right_max = minMax(root.right)
    return min(left_min, right_min, root.data), max(left_max, right_max, root.data)
    
```
